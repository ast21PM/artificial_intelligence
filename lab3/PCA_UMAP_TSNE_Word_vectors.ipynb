# Снижение размерности через PCA

<img src=https://files.speakerdeck.com/presentations/d57771b4a00540209f11c50be2dc8367/slide_111.jpg width=70%>

PCA docs  
https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```python
%%time

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# число главных компонент - размерноcть до которой хотим снизить исходную
n_components = 2
pca = PCA(n_components=n_components)

# снижаем размерность векторов
word_vectors_pca = pca.fit_transform(word_vectors)
word_vectors_pca.shape
```

```python
# каждое слово теперь не 25 а 2 мерный вектор
word_vectors_pca
```

Отрисовка результатов вариант 1 через bokeh

```python
import bokeh.models as bm
import bokeh.plotting as pl
from bokeh.io import output_notebook

output_notebook()
def draw_vectors(x, y, radius=10, alpha=0.25, color='blue', width=600, height=400, **kwargs):
    '''Рисует интерактивный график для точек данных с дополнительной информацией при наведении'''
    if isinstance(color, str):
        color = [color] * len(x)
    data_source = bm.ColumnDataSource({'x': x, 'y': y, 'color': color, **kwargs})
    fig = pl.figure(active_scroll='wheel_zoom', width=width, height=height)
    fig.scatter('x', 'y', size=radius, color='color', alpha=alpha, source=data_source)
    fig.add_tools(bm.HoverTool(tooltips=[(key, '@' + key) for key in kwargs.keys()]))
    pl.show(fig)
```

```python
draw_vectors(word_vectors_pca[:, 0], word_vectors_pca[:, 1], width=800, height=600, token=words)
```

Отрисовка результатов вариант 2 через plotly

```python
import plotly.express as px

# cоздание графика с использованием Plotly Express
fig = px.scatter(
    x=word_vectors_pca[:, 0],  # координаты X (первое измерение)
    y=word_vectors_pca[:, 1],  # координаты Y (второе измерение)
    hover_name=words,  # подсказки при наведении мыши (указать нужный столбец для отображения)
    title='Визуализация эмбеддингов',  # заголовок графика
    # color_continuous_scale=px.colors.sequential.Rainbow,  # цветовая схема для непрерывной шкалы
    # color_discrete_sequence=px.colors.qualitative.Dark24,  # цветовая схема для дискретной шкалы
)

# изменить размер точек
# fig.update_traces(marker=dict(size=3))

# настройка внешнего вида графика
fig.update_layout(
    xaxis_title='Dimension 1',  # подпись оси X
    yaxis_title='Dimension 2',  # подпись оси Y
    template='plotly_dark',  # применение темной темы
    # настройка размера графика
    autosize=False,
    width=1000,
    height=400,
    # применение темной темы
    # template='plotly_dark',
)
fig.show()
```

Как сделать чтобы интерактивные графики plotly занимали меньше памяти:
- отображать не каждую точку подряд а с шагом `step` (например каждую десятую)
- если тексты длинные то ограничить их макс длину перед визуализацией

```python
import plotly.express as px

# отображать каждую 10-ую точку чтобы график не занимал много памяти
step = 10

# cоздание графика с использованием Plotly Express
fig = px.scatter(
    x=word_vectors_pca[::step, 0],  # координаты X (первое измерение)
    y=word_vectors_pca[::step, 1],  # координаты Y (второе измерение)
    hover_name=words[::step],  # подсказки при наведении мыши
    title='Визуализация кластеров эмбеддингов',  # заголовок графика
    # color_continuous_scale=px.colors.sequential.Rainbow,  # цветовая схема для непрерывной шкалы
    # color_discrete_sequence=px.colors.qualitative.Dark24,  # цветовая схема для дискретной шкалы
)

# изменить размер точек
# fig.update_traces(marker=dict(size=3))

# настройка внешнего вида графика
fig.update_layout(
    legend_title_text='Кластеры',  # заголовок легенды
    xaxis_title='Dimension 1',  # подпись оси X
    yaxis_title='Dimension 2',  # подпись оси Y
    # настройка размера графика
    autosize=False,
    width=1000,
    height=400,
    # применение темной темы
    template='plotly_dark',
)
fig.show()
```

# Снижение размерности через UMAP

UMAP docs  
https://umap-learn.readthedocs.io/en/latest/  
https://github.com/lmcinnes/umap

```python
import umap
```

```python
%%time
# инициализация алгоритма и снижение размерности
# параметры UMAP: https://github.com/lmcinnes/umap?tab=readme-ov-file#how-to-use-umap
# увеличение n_neighbors делает акцент на глобальной структуре данных, но при этом может "размыть" локальные детали
word_vectors_umap = umap.UMAP(n_components=2, n_neighbors=15, random_state=111).fit_transform(word_vectors)
word_vectors_umap.shape
```

```python
import plotly.express as px

# cоздание графика с использованием Plotly Express
fig = px.scatter(
    x=word_vectors_umap[:, 0],  # координаты X (первое измерение)
    y=word_vectors_umap[:, 1],  # координаты Y (второе измерение)
    hover_name=words,  # подсказки при наведении мыши (указать нужный столбец для отображения)
    title='Визуализация кластеров эмбеддингов',  # заголовок графика
    color_continuous_scale=px.colors.sequential.Rainbow,  # цветовая схема
)

# настройка внешнего вида графика
fig.update_layout(
    legend_title_text='Кластеры',  # заголовок легенды
    xaxis_title='Dimension 1',  # подпись оси X
    yaxis_title='Dimension 2',  # подпись оси Y
    # настройка размера графика
    autosize=False,
    width=1000,
    height=400,
    # применение темной темы для графика
    template='plotly_dark',
)
fig.show()
```

# Снижение размерности через TSNE

TSNE docs  
https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html

```python
%%time

from sklearn.manifold import TSNE

# инициализация алгоритма и снижение размерности
tsne_model = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=3)
word_vectors_tsne = tsne_model.fit_transform(word_vectors)

word_vectors_tsne.shape
```

```python
from sentence_transformers import SentenceTransformer

# инициализация модели
embed_model = SentenceTransformer('cointegrated/rubert-tiny2')
```

```python
%%time

# извлечение слов из корпуса текста
words = np.array(model.index_to_key)[:2000]

# получение эмбедингов слов
embeddings = embed_model.encode(words)

# снизить размерность
embedding = umap.UMAP(n_neighbors=5).fit_transform(embeddings)
```

```python
# модель выдает эмбединги размерности 312
embeddings.shape
```

```python
import plotly.express as px

# cоздание графика с использованием Plotly Express
fig = px.scatter(
    x=embedding[:, 0],  # координаты X (первое измерение)
    y=embedding[:, 1],  # координаты Y (второе измерение)
    hover_name=words,  # подсказки при наведении мыши (указать нужный столбец для отображения)
    title='Визуализация кластеров эмбеддингов',  # заголовок графика
    color_continuous_scale=px.colors.sequential.Rainbow,  # цветовая схема
)

# настройка внешнего вида графика
fig.update_layout(
    legend_title_text='Кластеры',  # заголовок легенды
    xaxis_title='Dimension 1',  # подпись оси X
    yaxis_title='Dimension 2',  # подпись оси Y
    # настройка размера графика
    autosize=False,
    width=1000,
    height=400,
    # применение темной темы для графика
    template='plotly_dark',
)
fig.show()
```
